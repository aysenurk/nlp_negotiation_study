```python
from transformers import pipeline
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-mul")
translator(">>tur<< I've been waiting.")
```

```python
data["tr_translate"] = data.text.apply(lambda text: translator(">>tur<< " + text)[0]["translation_text"])
#translator([">>tur<< "+ text for text in data.text])
# data.text.apply(lambda text: translator(">>tur<< " + text)[0]["translation_text"])
```

```python
# Import generic wrappers
from transformers import AutoModel, AutoTokenizer 


# Define the model repo
model_name = "Helsinki-NLP/opus-mt-mul-en" 


# Download pytorch model
model = AutoModel.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)


# Transform input tokens 
inputs = tokenizer("Hello world!", return_tensors="pt")

# Model apply
outputs = model(**inputs)
    
```

```python
from transformers import MarianMTModel, MarianTokenizer


model_name = 'Helsinki-NLP/opus-mt-en-mul'

tokenizer = MarianTokenizer.from_pretrained(model_name)
print(tokenizer.supported_language_codes)

```

```python
model = MarianMTModel.from_pretrained(model_name)
```

```python
src_text = [
    '>>tur<< this is a sentence in english that we want to translate to french',
    '>>tur<< this is a sentence in english that we want to translate to french',
    '>>tur<< This should go to portuguese',
    '>>tur<< And this to Spanish'
]
translated = model.generate(**tokenizer(src_text, return_tensors="pt", padding=True))
[tokenizer.decode(t, skip_special_tokens=True) for t in translated]
```
